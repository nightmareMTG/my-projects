```python
import serpapi
```

```python
def search_flights(departure_airport, arrival_airport, departure_date, return_date):
    params = {
        "engine": "google_flights",
        "departure_id": departure_airport,
        "arrival_id": arrival_airport,
        "outbound_date": departure_date,
        "return_date": return_date,
        "currency": "USD",
        "hl": "en",
        "api_key": "74636d88363382283421c78c8170a33f0b675323b482e96d4f102a69593c3a23"
    }
    results = serpapi.search(params)
    return results
```

```python
departure_airport = input('Enter the departure airport IATA code')
arrival_airport = input('Enter the arrival airport IATA code')
departure_date = input('Enter the departure date in this format YYYY-MM-DD')
return_date = input('Enter the return date in this format YYYY-MM-DD')

search = search_flights(departure_airport, arrival_airport, departure_date, return_date)
```
